The Hierarchy (from lowest to highest)

DEBUG (10) -
Detailed information for debugging

INFO (20) -
Confirmation things are working

WARNING (30) -
Something unexpected but program still works

ERROR (40) -
Serious problem, function failed

CRITICAL (50) -
Program may crash, needs immediate attention

```python
import logging

#Configure the basic logging setting
logging.basicConfig(level=logging.DEBUG)

# log messages
logging.debug('This is a debug message')
logging.info('This is an info message')
logging.warning('This is a warning message')
logging.error('This is an error message')
logging.critical('This is a critical message')
```

```python
# Restart the kernel before running this cell

import logging

# Only show WARNING and above
logging.basicConfig(level=logging.ERROR)

logging.debug("This won't show")
logging.info("This won't show")
logging.warning("This will show")
logging.error("This will show")
logging.critical('This is a critical message')
```

```python
# Restart the kernel before running this cell
import logging

#configuring logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
    )

logging.debug("This won't show")
logging.info("This won't show")
logging.warning("This will show")
logging.error("This will show")
logging.critical('This is a critical message')
```
