```python
#Name all the columns that contain outliers
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']
for col in num_cols:
    print(f"\n{col}:")
    print(f"Min: {df[col].min()}")
    print(f"Q1: {df[col].quantile(0.25)}")
    print(f"Median: {df[col].median()}")
    print(f"Q3: {df[col].quantile(0.75)}")
    print(f"Max: {df[col].max()}")
    print(f"Std Dev: {df[col].std()}")
```

```python
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']
outliers = []
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    if ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).any():
        outliers.append(col)
print(outliers)
```
