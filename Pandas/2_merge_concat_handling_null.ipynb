<h1>Merging and Concatenating Data<h1>

```python
import pandas as pd
n=pd.read_csv("data/noc_regions.csv")
b = pd.read_csv("data/bios.csv")
```

```python
n.head()
```

```python
b.head()
```

![Joins](data/Joins.png)

```python
b_new=pd.merge(b,n,left_on='born_country',right_on='NOC',how='left')
```

```python
b_new.head()
```

```python
b_new.rename(columns={'region':'born_country_full'},inplace=True)
```

```python
b_new.head()
```

```python
b_new[b_new['NOC_x']!=b_new['born_country_full']][['name','NOC_x','born_country_full']]
```

```python
ind=b[b['born_country']=='IND'].copy()
usa=b[b['born_country']=='USA'].copy()
```

```python
ind_usa=pd.concat([ind,usa])
```

```python
ind_usa.head()
```

```python
ind_usa.tail()
```

<h2>Handling Null Values<h2>

```python
coffee = pd.read_csv("data/coffee.csv")
```

```python
coffee.info()
```

```python
import numpy as np
coffee.loc[[0,1],"Units Sold"]=np.nan # it make 0 and 1 index of unit sold to null value
```

```python
coffee.head()
```

```python
coffee.isna().sum()
```

```python
coffee.fillna(coffee['Units Sold'].median())
```

```python
#we can also drop the null values
coffee.dropna(subset=["Units Sold"]) #subset use if we want to drop null for specific columns
```

```python
coffee.head()
```

```python
coffee=coffee.dropna(subset=["Units Sold"])
```

```python
coffee.head()
```

```python
coffee.loc[[2,3],"Units Sold"]=np.nan
```

```python
coffee[coffee['Units Sold'].isna()]
```

```python
coffee[coffee['Units Sold'].notna()]
```
