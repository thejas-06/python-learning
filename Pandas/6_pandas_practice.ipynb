```python
def name(x):
    if x=="A":
        return "anil"
    elif x=="B":
        return "bhuvan"
    elif x=="C":
        return "chandan"
    elif x=="D":
        return "druv"
    elif x=="E":
        return "emily"
```

```python
df["full_name"]=df["Name"].apply(name)
```

```python
df
```

```python
df["Name"]=df["Name"].apply(name) #apply on same columns
```

```python
df
```

```python
df["Salary"].replace(50000,45000, inplace=True)
```

```python
df
```

```python
#convert all name in uppercase
for i in df["full_name"]:
    df["full_name"].replace(i,i.upper(), inplace=True)
```

```python
df
```

```python
# if salary is less then 20000 return low
# if salary is between 20000 and 40000 return mid
#else high
def level(x):
        if x<20000:
            return "low"
        elif 20000<=x<40000:
            return "mid"
        else:
            return "high"
```

```python
df["salary_level"]=df["Salary"].apply(level)
```

```python
df
```

```python
#length of each name in new column
for i in df["Name"]:
    print(len(i))
```

```python
def length(x):
    return len(x)
```

```python
df["len"]=df["Name"].apply(length)
```

```python
df["len1"]=df["Name"].apply(len)
```
