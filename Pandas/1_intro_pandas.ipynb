```python
bios["first_name"] = bios["name"].str.split().str[0]
```

```python
bios["first_name"].head()
```

```python
bios['born_date_time']=pd.to_datetime(bios["born_date"])
```

```python
bios.head()
```

```python
bios.info() #it look same but it changes object to datetime64[us]
```

```python
bios['born_year']=bios['born_date_time'].dt.year
bios[["name","born_year"]]
```

```python
bios['born_year']=bios['born_date_time'].dt.is_leap_year
bios[["name","born_year"]].head(10)
```

```python
bios['height_category']=bios['height_cm'].apply(lambda X: "Short" if X<165 else ('Average' if X<185 else "Tall"))
```

```python
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'LightWeight'
    elif row['height_cm']<185 and row['weight_kg']<=80:
        return 'MiddleWeight'
    else:
        'HeavyWeight'
bios["Category"]=bios.apply(categorize_athlete,axis=1)  #axis=1 means row '0' for columns
```
