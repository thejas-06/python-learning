```python
coffee=pd.read_csv("data/coffee.csv")
```

```python
coffee.head()
```

```python
coffee.groupby(['Coffee Type'])["Units Sold"].sum()
```

```python
coffee.groupby(['Coffee Type'])["Units Sold"].mean()
```

```python
coffee.groupby(['Coffee Type'])["Units Sold"].max()
```

```python
coffee.groupby(['Coffee Type'])["Units Sold"].min()
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold':'sum'})
```

```python
import numpy as np
coffee["price"]=np.where(coffee["Coffee Type"]=="Espresso", 150,200)
```

```python
coffee.head()
```

```python
coffee['revenue']=coffee['Units Sold']*coffee['price']
```

```python
coffee.head()
```

```python
pi=coffee.pivot(columns='Coffee Type',index='Day',values='revenue')
pi
```

```python
pi.loc['Monday','Latte'] #easy access
```

```python
pi.sum()
```

```python
pi.sum(axis=1)
```

```python
bios['born_date']=pd.to_datetime(bios['born_date'])
bios.groupby([bios['born_date'].dt.year]).count()
```

```python
bios.groupby([bios['born_date'].dt.year])['name'].count()
```

```python
bios.groupby([bios['born_date'].dt.year])['name'].count().reset_index()
```

```python
bios.groupby([bios['born_date'].dt.year])['name'].count().reset_index().sort_values('name')
```
