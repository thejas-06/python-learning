```python
import sqlite3
```

```python
#connect to an SQLite database
connection=sqlite3.connect("example.db")
connection
```

```python
cursor=connection.cursor()
```

```python
#create a table
cursor.execute('''
create table if not exists employees(
    id integer primary key,
    name text not null,
    age integer,
    department text
    )
''')

#commit the changes
connection.commit()
```

```python
#insert the data
cursor.execute('''
insert into employees(name,age,department)
               values('Thejas',20,'Data Scientist')
''')

cursor.execute('''
insert into employees(name,age,department)
               values('Teju',21,'Data Scientist')
''')

cursor.execute('''
insert into employees(name,age,department)
               values('Tezz',30,'Data Analyst')
''')

connection.commit()
```

```python
#query the data
cursor.execute('select * from employees')
rows=cursor.fetchall()

#print
for row in rows:
    print(row)
```

```python
#update the data in the table
cursor.execute('''
UPDATE employees
               Set age=39
               where name='Tezz'
''')
connection.commit()
```

```python
#query the data
cursor.execute('select * from employees')
rows=cursor.fetchall()

#print
for row in rows:
    print(row)
```

```python
#delete the data from the table
cursor.execute('''
Delete from employees
               where name='Tezz'
''')
connection.commit()
```

```python
#working with sales data
connection=sqlite3.connect('sales_data.db')
cursor=connection.cursor()

#create a table
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales(
    id INTEGER primary key,
    date text not NULL,
    product TEXT NOT NULL,
    sales integer,
    region text
    )
''')

#insert
sales_data=[
    ('2026-01-01','pro1',100,'south'),
    ('2026-01-02','pro2',200,'north'),
    ('2026-01-03','pro3',300,'east'),
    ('2026-01-04','pro4',400,'west'),
    ('2026-01-05','pro5',500,'south')
]

cursor.executemany('''
Insert into sales(date,product,sales,region)
                   values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
#close the connection
connection.close()
```
