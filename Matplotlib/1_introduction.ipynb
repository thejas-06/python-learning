<h1>Matplotlib.pyplot<h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [8,5,3,7,2,9]
plt.plot(x) # If we dont specify Y axis values it takes the index of X 
plt.show()
```

```python
plt.plot(x,"*")
plt.show()
```

```python
plt.plot(x,"*-")
plt.show()
```

```python
plt.plot(x,"*--")
plt.show()
```

```python
plt.plot(x,"*--",markersize=30,color="red")
plt.show()
```

```python
plt.plot(x,"--") # overwriting the *
plt.plot(x,"*")
plt.show()
```

```python
plt.plot(x,color="black")
plt.plot(x,"*",markersize=10,color="red") 
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,"--",linewidth=5,color="black")
plt.plot(x,"--",linewidth=2,color="white")
plt.plot(x,"o",markersize=20,color="yellow")
plt.plot(x,"*",markersize=10,color="red")
plt.show()
```

```python
x=np.random.randint(0,20, size=(5))
print(x)
y=np.random.randint(5,20 , size=5)
print(y)
```

```python
plt.plot(x,y,"o--")
for a,b in zip(x,y):
    plt.text(a,b,str(a))
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("graph")
plt.show()
```

Bar chart

```python
name =["a","b","c","d"]
score=[92,68,37,99]
```

```python
plt.bar(name,score,color=["purple","blue","red","green"])
plt.plot(score,"ko--",markersize=20) #k means black colour
plt.plot(score,"ro",markersize=10)
plt.plot(score,"ko",markersize=5)
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("name and score")
plt.show()
```

```python
z=plt.bar(name,score,color=["purple","blue","red","green"])
plt.bar_label(z)
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("name and score")
plt.show()
```

```python
k=[1,4,7,8,5,4,5,6,5]
plt.plot(k,"--",color="black")
plt.plot(k,"^",color="red",markersize=10)
plt.plot(k,"^",color="white",markersize=5)
plt.show()
```
