<h1>Line Graph<h1>

```python
df=pd.read_csv("data/gas_prices.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas prices over time(in USD)")
plt.plot(df.Year,df.USA,'b.-')
plt.plot(df['Year'],df['Japan'],'r.-')
plt.plot(df['Year'],df['South Korea'],'g.-')
plt.xticks(df.Year[::2])
plt.legend(["USA","Japan","South Korea"])
plt.xlabel("Year")
plt.ylabel("Price in USD")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas prices over time(in USD)")
for country in df:
    if country!="Year":
        plt.plot(df["Year"],df[country],marker=".")
plt.xticks(df.Year[::2])
plt.xlabel("Year")
plt.ylabel("Price in USD")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas prices over time(in USD)",fontdict={"fontweight":'bold','fontsize':14})
plt.plot(df.Year,df.USA,'b.-')
plt.plot(df['Year'],df['Japan'],'r.-')
plt.plot(df['Year'],df['South Korea'],'g.-')
plt.xticks(df.Year[::2].tolist()+[2011])
plt.legend(["USA","Japan","South Korea"])
plt.xlabel("Year")
plt.ylabel("Price in USD")
plt.show()
```

<h1>Pie Chart<h1>

```python
fifa.columns
```

```python
left=fifa.loc[fifa["Preferred Foot"]=="Left"].shape[0]
right=fifa.loc[fifa["Preferred Foot"]=="Right"].shape[0]
labels=['Left Foot', 'Right Foot']
colors=["#574f4b","#68ab9f"]
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f%%')
plt.title('Preferred Foot Distribution')
plt.show()
```

```python
left_count = len(fifa[fifa["Preferred Foot"] == "Left"])
right_count = len(fifa[fifa["Preferred Foot"] == "Right"])
plt.pie([left_count, right_count], labels=['Left Foot', 'Right Foot'], autopct='%1.1f%%')
plt.title('Preferred Foot Distribution')
plt.show()
```

```python
fifa.Weight.head()
```

```python
fifa.Weight=[int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight ]
```

```python
fifa.Weight.head()
```

```python
light=len(fifa[fifa.Weight<125])
light_medium=len(fifa[(fifa.Weight>=125) & (fifa.Weight<150)])
medium = len(fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)])
medium_heavy = len(fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)])
heavy = len(fifa[fifa.Weight >= 200])

plt.style.use("ggplot")
weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    
plt.show()
```
