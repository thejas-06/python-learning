```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,3,5],[6,2,4])
plt.show()
```

```python
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y)
plt.title("First")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y)
plt.title("First",fontdict={"fontname":"comic Sans MS","fontsize":20})
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=[0,1,2,2.5]
y=[0,3,4,5]
plt.plot(x,y,label="line",color="#abab",linewidth=3,marker="*",
         markersize=15,markeredgecolor="black",
         linestyle="--")
plt.title("First",fontdict={"fontname":"comic Sans MS","fontsize":20})
plt.xlabel("X")
plt.ylabel("Y")
plt.xticks([0,1,2,2.5,3])
plt.yticks([0,1,2,3,4,5])
plt.legend()
plt.show()
```

```python
x=[0,1,2,2.5]
y=[0,3,4,5]
#fmt =[color][marker][line]
plt.plot(x,y,'r.--',label="line")
plt.title("First",fontdict={"fontname":"comic Sans MS","fontsize":20})
plt.xlabel("X")
plt.ylabel("Y")
plt.xticks([0,1,2,2.5,3])
plt.yticks([0,1,2,3,4,5])
plt.legend()
plt.show()
```
