```python
import qrcode
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data('https://github.com/thejas-06')
qr.make(fit=True)

img = qr.make_image(fill_color="purple", back_color="white")

img.save("38_MyGitHub.png")
```
