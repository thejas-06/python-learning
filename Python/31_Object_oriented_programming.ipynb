<h1> Object oriented programming <h1>

```python
#syntax for constractor
class ClassName:
    def __init__(self, parameter1, parameter2):
        self.attribute1 = parameter1
        self.attribute2 = parameter2
```

```python
class human:
    #attributes
    def __init__(self,name):
        self.name=name  #instance variable

    #method
    def walk(self):
        print(f"{self.name} is walking")

    def learn(self):
        print(f"{self.name} is learning new things")

#creating an object of class
tej=human("Thejas")

tej.walk()
tej.learn()
```

```python
class mobile:
    def __init__(self,brand,price):
        self.b=brand
        self.p=price

    def dis(self):
        print(f"mobile brand is {self.b} and it cost you about {self.p}Rs without charger")

m=mobile("Nothing Phone 1" , 30000)
m.dis()
```

```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def display_info(self):
        print(f"{self.name} and markes is {self.marks}")
s1=student("thejas",90)
s2=student("tejas",80)
s3=student("teju",70)
s4=student("tezz",60)

s1.display_info()
s2.display_info()
s3.display_info()
s4.display_info()
```

```python
# Self is just a naming convention We can use any name
class student:
    def __init__(nanu,name,marks):
        nanu.name=name
        nanu.marks=marks
    def display_info(nanu):
        print(f"{nanu.name} and markes is {nanu.marks}")
s1=student("thejas",90)
s2=student("tejas",80)
s3=student("teju",70)
s4=student("tezz",60)

s1.display_info()
s2.display_info()
s3.display_info()
s4.display_info()
```

```python
#Optional Parameters in Constructors
class Book:
    def __init__(self, title, author="Unknown"):
        self.title = title
        self.author = author

    def show_book(self):
        print(f"Title: {self.title}, Author: {self.author}")

book1 = Book("Python Programming")
book2 = Book("Machine Learning", "Andrew Ng")

book1.show_book()
book2.show_book()
```

```python
class employee:
    def __init__(self, name, salary=30000): #default value
        self.name=name
        self.salary=salary
    
    def dis(self):
        print(f"name {self.name} and salary {self.salary}")

e1=employee("tejas",90000)
e2=employee("tezz")

e1.dis()
e2.dis()
```
