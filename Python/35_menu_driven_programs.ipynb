```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b

def menu():
    print("\n")
    print("***********Simple Calculator***********")
    print("enter your choice")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Quit\n")
    
while (True):
    menu()

    choice=int(input("enter your choice: "))
    if choice in {1,2,3}:
        a=int(input("enter first number: "))
        b=int(input("enter second number: "))
    if choice==1:
        print("      addition is",add(a,b))
    elif choice==2:
        print("subtraction is",sub(a,b))
    elif choice==3:
        print("multiplication is",mul(a,b))
    elif choice==4:
        print("quiting..")
        break
    else:
        print("Invalid choice. try again")
```

```python
#banking system
class Account:
    def __init__(self,id,holder_name):
        self.id=id
        self.holder_name=holder_name
        self._balance=0 #encapsulation

    def check_balance(self):
        print(f"balance:{self._balance}")

    def deposit(self,amount):
        self._balance+=amount
        print(f"deposit successful. updated: {self._balance}")

    def withdraw(self,amount):
        if self._balance>=amount:
            self._balance-=amount
            print(f"withdraw successful.updated balance: {self._balance}")
        else:
            print("Not enough money")

class SavingAccount(Account): #inheritance
    def calculate_interest(self):
        interest_rate=0.04
        interest=self._balance * interest_rate
        print(f"interest: {interest}")

class CurrentAccount(Account):
    def withdraw(self,amount): # Polymorphism
        OverDraft_Limit= 1000
        if self._balance+OverDraft_Limit>=amount:
            self._balance-=amount
            print(f"withdraw successful.updated balance: {self._balance}")
        else:
            print("ask is over limit")

class Bank:
    def __init__(self,name,city):
        self.name=name
        self.city=city
        self.__accounts={}

    def create_account(self,id,holder_name,type):
        if type=="savings":
            new_account=SavingAccount(id,holder_name)
        elif type=="current":
            new_account=CurrentAccount(id,holder_name)
        else:
            print("Invalid account type")
            return None
        
        self.__accounts[id]=new_account
        print("account creation successful")
        return new_account
    
    def get_account(self,id):
        if id not in self.__accounts:
            print("account not found")
            return None
        else:
            account = self.__accounts[id]
            print(f"\nID:{account.id}\nHolder name: {account.holder_name} ")
            return account
        
tbk=Bank("tejas bank of karnataka","musuru")

s1=tbk.create_account("1","thejas","savings")
c1=tbk.create_account("2","virat","current")

s1.deposit(1000)
c1.deposit(100)

s1.withdraw(2000)
c1.withdraw(200)

s1.calculate_interest()
```

```python
#Student Report Card Generator
class Student:
    def __init__(self,roll_no,name):
        self.name=name
        self.roll_no=roll_no
        self.__marks={}

    def get_marks(self):
        return self.__marks

    def add_marks(self,subject,marks):
        self.__marks[subject]=marks

    def calculate_average(self):
        if not self.__marks:
            return 0
        total = sum(self.__marks.values())
        return total / len(self.__marks)


    def is_pass(self):
        for mark in self.__marks.values():
            if mark<35:
                print(f"{self.name} is fail")
                return
        print(f"{self.name} is pass")

    def calculate_grade(self):
        print("Grade: ", end="")
        percentage = self.calculate_average()
        if percentage>=90:
            print("A")
        elif percentage>=85:
            print("B")
        else:
            print("c")

class ReportCard:
    @staticmethod
    def generate(student:Student):
        student_marks = student.get_marks()
        print(f"\nName: {student.name} \t Roll No. {student.roll_no}")
        print("------marks-------")
        for subject, marks in student_marks.items():
            print(f"{subject} = {marks}")
        print("----------------------")
        print(f"Average: {student.calculate_average()}") #with return
        student.is_pass() # without return
        student.calculate_grade() # without return
        print("\n")

s=Student(1, "Thejas")
s.add_marks("Math", 90)
s.add_marks("Science", 36)

ReportCard.generate(s) # only for static method
```
