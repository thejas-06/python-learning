List

```python
l = [4, 7, 2, 8, 2, 9]
print("Original list:", l)

# Append 5 to the end
l.append(5)
print("After append(5):", l)

# Insert 9 at index 4
l.insert(4, 9)
print("After insert(4, 9):", l)

# Extend with multiple elements
l.extend([88, 45])
print("After extend([88, 45]):", l)

# Remove first occurrence of 2
l.remove(2)
print("After remove(2):", l)

# Sort in descending order
l.sort(reverse=True)
print("After sort:", l)

# Remove element at index 2
popped = l.pop(2)
print(f"After pop(2) (removed {popped}):", l)

# Count occurrences of 2
con = l.count(2)
print(f"Count of 2 in list: {con}")
print("List remains:", l)

# Clear all elements
l.clear()
print("After clear():", l)
```

```python
l =[7,8,4,5,[55,88,99,[100,200,300,400],550]]
# 1. Remove 400 from the list
li = l[4][3]
print(li)
li.remove(400)
print(li)
# 2. Add 1000 on 3rd index
l.insert(3,1000)
print(l)
# 3. Extract 5 to 550 from the list
lis = l[4:]
print(lis)
# 4. Replace 300 with 3000
l[5][3][2]=3000
print(l)
```
