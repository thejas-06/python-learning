<h1>Getters and Setters<h1>

```python
#without
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

s=Student("thejas",20)

print(s.name)
print(s.age)

s.age=100 # no controlled access
print(s.age)
s.age="20"
print(s.age)
```

```python
#with (they help in accidental modification,protect,validate,controlled)
class Student:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age

    def get_name(self): #getter
        return self.__name

    def set_name(self,name): #setter
        self.__name=name

    def set_age(self,age):
        if isinstance(age,int):
            self.__age=age
        else:
            print("Age should be int")

s=Student("thejas",20)

print(s.get_name())
s.set_name("teju")
print(s.get_name())
s.set_age("20")
```

<h1>Method Overloading<h1>

```python
#Python doesnâ€™t support method overloading directly, but we can achieve it by using default parameters
class Calculator:
    def add(self,a,b,c=0):
        print(a+b+c)

c=Calculator()
c.add(2,4)
c.add(2,4,6)
```

<h1>Method Overriding<h1>

```python
class Animal:
    def sound(self):
        print("animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")  # Overrides the parent class method

animal = Animal()
animal.sound()
dog = Dog()
dog.sound()  # Calls the overridden method in Dog class
```

<h1>Super Function<h1>

```python
class Animal:
    def sound(self):
        print("animal sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog barks")

dog = Dog()
dog.sound() 
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)  # Calling the parent class's __init__ method

    def sound(self):
        super().sound()  # Calling the parent class's sound method
        print(f"{self.name} barks")

    def get_angry(self):
        super().sound()
        print("bow..")

dog = Dog("doggy")
dog.sound()
dog.get_angry()
```

<h1>Abstract Class<h1>

```python
"""An abstract class in Python is a class that cannot be instantiated directly. 
It can have abstract methods, which must be implemented by subclasses."""

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class bike(Vehicle):
    def __init__(self,name):
        self.name=name
    def start_engine(self):
        print("starting engine..")

b=bike("Royal")
print(b.name)
```
