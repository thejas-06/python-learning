```python
class Movie:
    def __init__(self,title,rating):
        self.title=title
        self.rating=rating

    def display(self):
        print(f"Title {self.title} and rating is {self.rating}")

m=Movie("KGF",8.5)
m.display()
```

```python
#Encapsulation
class bankAccount:
    def __init__(self,acc_no,balance):
        self.__acc_no=acc_no
        self.__balance=balance

    def check_balance(self):
        print(self.__balance)

    def deposit(self,amount):
        self.__balance+=amount
    
    def withdraw(self,amount):
        if self.__balance<amount:
            print("insufficient funds")
            return
        self.__balance-=amount
        print(f"withdraw successful-balance:{self.__balance}")

a=bankAccount(acc_no=123,balance=10,)
a.deposit(100)
a.withdraw(500)
a.check_balance()    
```

```python
#abstraction
class phone:
    def take_picture(self):
        print("picture taken")

    def call_contact(self):
        print("calling..")

p=phone()
p.call_contact()
p.take_picture()
```

```python
#inheritance
class Vehicle:
    def start(self):
        print("starting..")

class bike(Vehicle):
     def __init__(self,name):
         self.name=name

     def ride(self):
         print(f"{self.name} ready to ride")

b=bike("Hero")
b.start()
b.ride()
```

```python
#polymorphism
class shape:
    def calc_area(self):
        print("Area calculated")

class circle(shape):
    def __init__(self,radius):
        self.radius=radius

    def calc_area(self):
        print(f"area of circle {3.14*self.radius**2}")

class rectangle(shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b

    def calc_area(self):
        print(f"area of rectangle {self.l*self.b}")


c=circle(2)
r=rectangle(2,3)
c.calc_area()
r.calc_area()
```

```python
#getters and setters
class BankAccount:
    def __init__(self,balance):
        self.__balance=balance

    def get_balance(self):
        return self.__balance

    def set_balance(self,updated_balance):
        if updated_balance<0:
            print("Error")
            return
        self.__balance=updated_balance

b=BankAccount(100)
print(b.get_balance())
b.set_balance(1000)
b.get_balance()
```

```python
#method overloading
class Calculator:
    def multiply(self,a,b,c=1):
        print(a*b*c)

c=Calculator()
c.multiply(3,2)
c.multiply(2,2,2)
```

```python
#method overriding
class shape:
    def draw(self):
        print("Drawing shape")

class Circle(shape):
    def draw(self):
        super().draw()
        print("Drawing circle")

c=Circle()
c.draw()
```

```python
#abstract classes
from abc import ABC, abstractmethod

class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass

class manager(employee):
    def calculate_salary(self):
        print("manager salary is calculated")

m=manager()
m.calculate_salary()
```
