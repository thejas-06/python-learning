```python
num =int(input("Enter any number: "))
match num:
    case 1:
        print("One")
    case 2:
        print("two")
    case 3:
        print("three")
    case _:
        print("not in range")
```

```python
light = input("Enter traffic light color (red/yellow/green): ").lower()

match light:
    case "red":
        print("STOP! Wait for green light.")
    case "yellow":
        print("SLOW DOWN! Prepare to stop.")
    case "green":
        print("GO! You can proceed.")
    case _:
        print("Invalid traffic light color!")
```

```python
print("Calculator")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator: ")
match op:
    case "+":
        result = num1 + num2
    case "-":
        result = num1 - num2
    case "*":
        result = num1 * num2
    case "/":
        result = num1 / num2
    case _:
        result = "invalid operator"

print(f"Result: {result}")
```

```python
time =17
is_hungry=False
match time:
    case 9:
        print("Breakfast")
    case 14 | 13:
        print("Lunch")
    case 17 if is_hungry:
        print("Snacks")
    case _:
        print("Work")
```
