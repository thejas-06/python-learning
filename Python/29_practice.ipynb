```python
#write a function to calculate 2nd hightest element in a list
def second_hightest(p):
    p.sort()
    f=0
    for i in p:
        if i>min(p) and i<max(p):
            f=i
    return f
p = [1,4,7,8,2,7,8,3,8]
second_hightest(p)
```

```python
# write a function to print only string element from the list
def only_string(st):
    for i in st:
        if isinstance(i, str):
            print(i)

st = [1, 4, "thejas", 9, 2, 88, "we"]
only_string(st)
```
