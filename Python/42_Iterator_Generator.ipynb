<h1>Iterators<h1>

```python
nums=[23,65,87]
it=iter(nums)
#iterating here or traversing
print(next(it))
```

```python
nums=[23,65,87]
it=iter(nums) #return the iterator object
print(next(it)) #return the next item from the sequence
print(next(it))
print(next(it))
```

```python
class CountDown:
    def __init__(self, start):
        self.start = start
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.start <= 0:
            raise StopIteration
        num = self.start
        self.start -= 1
        return num

cd = CountDown(5)
for i in cd:
    print(i)
```

<h1>Generators<h1>

```python
def f(x):
    yield x*2

g=f(5)
for i in g:
    print(i)
```

```python
def gen():
    yield 1
    yield 2
    yield 3
    
g=gen()
for i in g:
    print(i)
```

```python
def gen():
    yield 1
    yield 2
    yield 3
    
g=gen()
print(next(g))
print(next(g))
# print(next(g))
```

```python
def gen(x):
    for i in range(x):
        yield i
    
g=gen(6)
for i in g:
    print(i)
```

```python
def gen(x):
    for i in range(x):
        yield i
c=0
g=gen(6)
for i in g:
    if c>2:
        break
    print(i)
    c+=1
```

<h1>Generator Expressions<h1>

```python
#without generator expression
#-it is a list
#-it not memory efficient
#List stores all elements in memory
import sys
l=[x*x for x in range(1,100)]
# print(l)
print(type(l))
print(sys.getsizeof(l))
```

```python
#Generator just stores the generator object itself
gl=(x*x for x in range(1,11))
print(type(gl))
# print(next(gl))
print(sys.getsizeof(gl))
```
