```python
#without
def decorator_name(func):
    def wrapper():
        print("Hi!")
        func()
        print("Thejas")
    return wrapper

def intro():
    print("My name is:")

intro()
```

```python
#without
def decorator_name(func):
    def wrapper():
        print("Hi!")
        func()
        print("Thejas")
    return wrapper
@decorator_name
def intro():
    print("My name is:")
@decorator_name
def by():
    print("byeeee.")

intro()
print("\n")
by()
```

```python
def show_result(func):
    def wrapper(a,b):
        print("Result: ", end="")
        func(a,b)
    return wrapper
@show_result
def add(a,b):
    print(a+b)

add(2,3)
```

```python
#Logs
def logger(func):
    def wrapper():
        print(f"Function '{func.__name__}' is being called.")
        func()
    return wrapper

@logger
def greet():
    print("Hey there!")

greet()
```

```python
def login_required(func):
    def wrapper():
        print("Checking if user is logged in...")
        func()
    return wrapper

@login_required
def view_profile():
    print("thejas profile opened.")

view_profile()
```
