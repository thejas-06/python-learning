Map: transform each element

```python
nums=[1,2,3,4,5]
def double(i):
    return i*2
res=map(double,nums)
print(list(res))
```

```python
nums=[1,2,3,4,5]
res=map(lambda i:i*2,nums)
print(list(res))
```

```python
a=[1,2,3,4]
b=[5,6,7,8]
res=map(lambda x,y:x+y,a,b)
print(list(res))
```

filter: select elements based on condition

```python
nums=[1,2,3,4,5,6]
def is_even(i):
    return i%2==0
res=filter(is_even,nums)
print(list(res))
```

```python
nums=[1,2,3,4,5,6]
res=filter(lambda i:i%2!=0,nums)
print(list(res))
```

```python
marks=[19,40,34,70,90]
result=filter(lambda m:m<36,marks) 
print(list(result))
```

Reduce: reduce iterable to single value

```python
from functools import reduce
num=[1,2,3,4]
res=reduce(lambda x,y:x+y,num)
print(res)
```

```python
num=[25,76,93,12,54,76]
result=reduce(lambda a,b:a if a>b else b,num)
print(result)
```

```python
from functools import reduce
scores = [45, 67, 89, 19, 76, 30]

# 1. Increase all scores by 5 using map
updated = list(map(lambda x: x + 5, scores))

# 2. Filter only passing students (>= 35)
passed = list(filter(lambda x: x >= 35, updated))

# 3. Find the total marks of all passed students using reduce
total = reduce(lambda x, y: x + y, passed)

print("Updated Scores:", updated)
print("Passed Students:", passed)
print("Total Marks:", total)
```
