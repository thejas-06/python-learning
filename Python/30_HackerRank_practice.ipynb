```python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    avg_score = sum(student_marks[query_name])/3
    print("{:.2f}".format(avg_score))

'''sample input:
2
Harsh 25 26.5 28
Anurag 26 28 30
Harsh
'''
```

```python
# In a classroom of N students. find the student with the second lowest grade

if __name__ == '__main__':
    records=[]
    s=set()
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name , score])
        s.add(score)
    sec_low_score=sorted(s)[1]  #sorted function automatically converts to list
    sec_low_name=[]
    
    for name,score in records:
        if score==sec_low_score:
            sec_low_name.append(name)
            
    for name in sorted(sec_low_name):
        print(name)

"""Sample Input
5
Harry
37.21
Berry
37.21
Tina
37.2
Akriti
41
Harsh
39"""
```
