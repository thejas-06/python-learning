```python
def accu(income):
    taxable_income=income-1200000
    return taxable_income

def income_tax(income):
    if income<=1200000:
        print("no tax")
    elif 1200000<income<=1500000:
        taxable=accu(income)
        tax = taxable*(5/100)
        print(tax)
    elif 1500000<income<=2000000:
        taxable=accu(income)
        tax = taxable*(10/100)
        print(taxable)
    else:
        taxable=accu(income)
        tax = taxable*(20/100)
        print(taxable)
income_tax(1600000)
```

```python
def income_tax(income):
    if income <= 1200000:
        print("no tax")
    elif income <= 1500000:
        # Only on amount above 12L
        tax = (income - 1200000) * 0.05
        print(f"Tax: ₹{tax}")
    elif income <= 2000000:
        # 5% on first 3L above 12L + 10% on rest
        tax = 15000 + (income - 1500000) * 0.10
        print(f"Tax: ₹{tax}")
    else:
        # 5% on 3L + 10% on 5L + 20% on rest
        tax = 15000 + 50000 + (income - 2000000) * 0.20
        print(f"Tax: ₹{tax}")

income_tax(1300000)  # Output: Tax: ₹5000.0
```
